```python
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)


# Plot X versus X/2 on the top left axes


# Plot a scatter plot of 10 random numbers on each axis on the top right subplot


# Plot a bar graph of the favourite food keys and values on the bottom left subplot


# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot

import numpy as np
import matplotlib.pyplot as plt

# Prepare data
X = np.linspace(0, 10, 100)  # For line plot

# For scatter plot
x_scatter = np.random.rand(10)
y_scatter = np.random.rand(10)

# For bar plot
favorite_foods = {
    'Pizza': 12,
    'Sushi': 18,
    'Burger': 10
}
foods = list(favorite_foods.keys())
prices = list(favorite_foods.values())

# For histogram
normal_data = np.random.randn(1000)

# Create 2x2 subplots with figsize (10, 5)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
fig.suptitle('2x2 Subplots Example', fontsize=16)

# Top left: Line plot (X vs X/2)
axes[0, 0].plot(X, X/2)
axes[0, 0].set_title('Line Plot: X vs X/2')

# Top right: Scatter plot of random numbers
axes[0, 1].scatter(x_scatter, y_scatter)
axes[0, 1].set_title('Scatter Plot: Random Points')

# Bottom left: Bar plot of favourite foods
axes[1, 0].bar(foods, prices)
axes[1, 0].set_title('Bar Plot: Favourite Foods')

# Bottom right: Histogram of normal distribution
axes[1, 1].hist(normal_data, bins=30, edgecolor='black')
axes[1, 1].set_title('Histogram: Normal Dist.')

# Adjust layout
plt.tight_layout()
fig.subplots_adjust(top=0.88)  # Leave space for main title

# Show the plot
plt.show()

```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

Here 'Price' column is still stored as text (object dtype) because it contains symbols like $ and ,.
So we cant plot it as numeric data until we clean and convert it!

Heres how to fix it properly:
Well:
1) Clean 'Price' so its numeric
2) Add a new column 'Total Sales' using .cumsum()
3) Add a 'Sale Date' column using pd.date_range()

```python
# Clean 'Price' column: remove $, commas, last 2 digits, then convert to int
car_sales['Price'] = car_sales['Price'].astype(str).replace(r'[$,]', '', regex=True).str[:-2].astype(int)
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame

import pandas as pd

# Assuming you already have the cleaned 'Price' column:
# car_sales['Price'] = car_sales['Price'].replace(r'[$,]', '', regex=True).str[:-2].astype(int)

# 1 Add 'Total Sales' column (cumulative sum of 'Price')
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# 2 Add 'Sale Date' column (successive dates starting today)
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(car_sales))

# 3 View the updated DataFrame
print(car_sales.head())
```

```python
# Replicate the above plot in whichever way you see fit

# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old


# Create the plot


# Plot the data


# Customize the plot


# Add a meanline


import pandas as pd
import matplotlib.pyplot as plt

# 1 Create DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease['age'] > 50]

# 2 Create the plot and figure
fig, ax = plt.subplots(figsize=(10, 6))

# 3 Plot the data: scatter plot with color by target
scatter = ax.scatter(
    over_50['age'],
    over_50['chol'],
    c=over_50['target'],  # color by target
    cmap='bwr'  # blue-white-red color map
)

# 4 Customize the plot
ax.set_title('Heart Disease: Age vs Cholesterol (Patients Over 50)')
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')

# Add color bar for target
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Heart Disease (0 = No, 1 = Yes)')

# 5 Add a mean line for cholesterol
chol_mean = over_50['chol'].mean()
ax.axhline(chol_mean, linestyle='--', color='grey', label=f'Mean Cholesterol ({chol_mean:.0f})')

# Add legend for mean line
ax.legend()

plt.show()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like plot3

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot


# Plot the data


# Customize the plot


# Add a meanline

import matplotlib.pyplot as plt

# Use the desired style
plt.style.use('seaborn-v0_8-whitegrid')  # or 'seaborn-whitegrid'

# Filter patients over 50
over_50 = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data with cmap='winter'
scatter = ax.scatter(
    over_50['age'],
    over_50['chol'],
    c=over_50['target'],
    cmap='winter'  # New colormap
)

# Customize the plot
ax.set_title('Heart Disease: Age vs Cholesterol (Patients Over 50)')
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')

# Add color bar for target
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Heart Disease (0 = No, 1 = Yes)')

# Add meanline with color='red'
chol_mean = over_50['chol'].mean()
ax.axhline(chol_mean, linestyle='--', color='red', label=f'Mean Cholesterol ({chol_mean:.0f})')

# Add legend
ax.legend()

# Show the plot
plt.show()
```
